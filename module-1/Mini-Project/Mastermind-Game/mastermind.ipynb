```python
import re
def mastermind_game():
        colors = ['red', 'green', 'blue', 'yellow', 'white', 'brown' ]
        number_of_rounds = 0
        
        
        def even_number():
            number = int(input("Enter the number of rounds to play: "))
            if number%2 == 0:
                return number
            else:
                return even_number()

            
        def get_valid_codebreaker():
            codebreaker_input = input("Choose any four colors from the above colors list separated by commas: \n ")
            codebreaker_list = re.split('[ ,]+', codebreaker_input)
            for color in codebreaker_list:
                if color not in colors:
                    return get_valid_codebreaker()
            
            
            codebreaker_tuple = tuple(codebreaker_list)
            return codebreaker_tuple
                                      
            
        number_of_rounds = even_number()
#         print(f"Number of rounds to play are: {number_of_rounds}")

        while number_of_rounds > 0:
            import random
            guess = 1
            max_guesses = 2
            
            # 
            set1 = set()
            while (len(set1) < 4):
                set1.add(random.choice(colors)) 

            codemaker = tuple(set1)

            while(guess <= max_guesses):
                keys_peg = []
                print(colors)
                
                codebreaker = get_valid_codebreaker()
                if codemaker == codebreaker:
                    print(f"Congratulations codebreaker won the game in guess {guess}.\n")
                    break
                else:
                    for item in codebreaker:
                        if item in codemaker:
                            if codebreaker.index(item) == codemaker.index(item):
                                keys_peg.append("red")
                            else:
                                keys_peg.append('white')
                        else:
                            keys_peg.append('blank')

                keys_peg.sort()
                print(f"Guess:{guess} {keys_peg}\n")

                if guess == max_guesses:
                    print("Congratulations codemaker won the game.\n")

                guess += 1
            
            
            number_of_rounds -= 1
            if(number_of_rounds != 0):
                print("****Starting New Game****")
            

mastermind_game()
```
