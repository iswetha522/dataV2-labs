#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
# your code here
html = requests.get(url).content
html
soup = BeautifulSoup(html, 'lxml')
soup
tags = ['a']
links = soup.find_all('a')

urls = []

for link in links:
    if 'href' in link.attrs:
        url = link.attrs['href']
#         urls.append(url)
        if url.startswith('https:'):
            urls.append(url)
        elif url.startswith('//'):
            urls.append('https:' + url)
        elif url.startswith('/w') or url.startswith('/wiki'):
            urls.append('https://en.wikipedia.org' + url)
        else:
            continue

urls
````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
# your code here
html = requests.get(url).content
html
soup = BeautifulSoup(html, 'lxml')
soup
cdict = {'class' : 'usctitlechanged'}
titles = soup.find_all('div', cdict)
text = [element.text for element in titles]
text
lst = []
for row in text:
    if (row.startswith('\n\n          ') and row.endswith('\n\n        ')) or (row.startswith('\n\n          ') and row.endswith('\n')) :
        new_row = row.strip().replace('\n', '')
        lst.append(new_row)
    
lst
````

####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
# your code here
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')

selector = 'div#content > table > tbody > tr'

rows = []

for row in soup.select(selector):
    
    # date, longiture, latitude, region
    col = []
    
    date = row.select('td:nth-child(4) > b > a')[0].text
    col.append(date)
    
    longitude = row.select('td:nth-child(5)')[0].text
    longitude_direction = row.select('td:nth-child(6)')[0].text
    col.append(longitude + ' ' + longitude_direction)
    
    
    latitude = row.select('td:nth-child(7)')[0].text
    latitude_direction = row.select('td:nth-child(8)')[0].text
    col.append(latitude + ' ' + latitude_direction)
    
    region = row.select('td:nth-child(11)')[0].text
    col.append(region)
    
    rows.append(col)
    
colnames = ['Date', 'Logitude', 'Latitude', 'Region']
df = pd.DataFrame(rows, columns = colnames)
df.head(20)
````
