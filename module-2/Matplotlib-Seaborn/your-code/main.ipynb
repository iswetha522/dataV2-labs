#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100)
y = x ** 2
z = 2.718 ** x
fig, [ax1, ax2] = plt.subplots(1,2, figsize = (12,6))
ax1.plot(x,y)
ax1.set_title('Normal Plot')
ax1.set_xlabel('x linear')
ax1.set_ylabel('y linear')
ax1.set_xscale('linear')
ax1.set_yscale('linear')

ax2.plot(x,z)
ax2.set_title('Exponential Plot')
ax2.set_xlabel('x logarithmic ')
ax2.set_ylabel('y logarithmic ')
ax2.set_xscale('linear')
ax2.set_yscale('log')

fig.suptitle('figure with subplots')
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100)
y = x ** 2
z = 2.718 ** x
fig, [ax1, ax2] = plt.subplots(1,2, figsize = (12,6))
l1 = ax1.plot(x,y, color="red")
ax1.set_title('Normal Plot')
ax1.set_xlabel('x linear')
ax1.set_ylabel('y linear')
ax1.set_xscale('linear')
ax1.set_yscale('linear')

l2 = ax2.plot(x,z, color="blue")
ax2.set_title('Exponential Plot')
ax2.set_xlabel('x linear ')
ax2.set_ylabel('y logarithmic ')
ax2.set_xscale('linear')
ax2.set_yscale('log')

line_labels = ["Normal", "Logarithmic"]
fig.suptitle('figure with subplots')
fig.legend([l1, l2],     # The line objects
           labels=line_labels,   # The labels for each line
           loc="center right",   # Position of legend
           borderaxespad=1,    # Small spacing around legend box
           title="Legend Title"  # Title for the legend
           )
plt.subplots_adjust(right=0.85)
plt.show()
```

#### From the Fitbit data, we want to visually understand:

```python
import seaborn as sns
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_month_encoded = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()
median_month_encoded
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_work_or_weekend = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()
median_work_or_weekend
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12,4)) 
plt.sca(ax[0]) 
median_month_encoded.plot() 
months = [' ','Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(13), months, rotation = 0)
  
plt.sca(ax[1]) 
median_work_or_weekend.plot(kind = 'bar') 
plt.xticks(range(2),['Weekend', 'Weekdays'], rotation = 0)
  
     
plt.show()
```
